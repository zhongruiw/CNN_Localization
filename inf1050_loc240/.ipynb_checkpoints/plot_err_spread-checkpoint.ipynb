```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

obs_freq = 50
obs_dens = 4
obs_grid_plt = 0
pltlenth = 50
pltbegin = 0
pltend = pltbegin + pltlenth
truth = np.load('truth_times.npy')[pltbegin*obs_freq:pltend*obs_freq, (obs_grid_plt+1)*obs_dens-1]
obs = np.load('observations.npy')[pltbegin:pltend, obs_grid_plt]
model  =  np.load('zens_times_prior.npy')[pltbegin*obs_freq:pltend*obs_freq, (obs_grid_plt+1)*obs_dens-1]
analy  =  np.load('zeakf_analy.npy')[(obs_grid_plt+1)*obs_dens-1, pltbegin:pltend]
# zeakf_analy_inde is obtained by applying kg_t to basic ensemble
analy_indeself  =  np.load('zeakf_analy_indeself.npy')[(obs_grid_plt+1)*obs_dens-1, pltbegin:pltend] 

times = np.linspace(pltbegin*obs_freq, (pltend-1)*obs_freq, pltlenth*obs_freq)
assim_times = np.linspace(pltbegin*obs_freq, (pltend-1)*obs_freq, pltlenth)
plt.plot(times, truth, 'k', label='truth')
plt.plot(times, model, 'b', label='prior')
plt.plot(assim_times, analy, 'r', label='analysis')
plt.plot(assim_times, analy_indeself, 'r-.', label='analy_indeself')
plt.plot(assim_times, obs, 'g*', label='observation')

plt.legend(loc='upper left')
plt.xlabel('time step')
plt.ylabel('x at grid0')

plt.show()
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

time_steps = 200*360
zt = np.load('truth_times.npy')[0: time_steps + 1: obs_freq, :].T
# zeakf_prior  =  np.load('zeakf_prior.npy')
# zeakf_prior_inde  =  np.load('zeakf_prior_inde.npy')
zeakf_analy  =  np.load('zeakf_analy.npy')
# zeakf_analy_inde  =  np.load('zeakf_analy_inde.npy')
zeakf_analy_indeself  =  np.load('zeakf_analy_indeself.npy')
prior_spread_rmse = np.load('prior_spread_rmse.npy')
prior_spread_inde_rmse = np.load('prior_spread_inde_rmse.npy')
analy_spread_rmse = np.load('analy_spread_rmse.npy')
analy_spread_inde_rmse = np.load('analy_spread_inde_rmse.npy')

iobsbeg = 41
iobsend = int((time_steps/obs_freq)) + 1

# prior_rmse = np.sqrt(np.mean((zt - zeakf_prior) ** 2, axis=0))
# prior_inde_rmse = np.sqrt(np.mean((zt - zeakf_prior_inde) ** 2, axis=0))
analy_rmse = np.sqrt(np.mean((zt - zeakf_analy) ** 2, axis=0))
# analy_inde_rmse = np.sqrt(np.mean((zt - zeakf_analy_inde) ** 2, axis=0))
analy_indeself_rmse = np.sqrt(np.mean((zt - zeakf_analy_indeself) ** 2, axis=0))

# prior_err = np.mean(prior_rmse[iobsbeg - 1: iobsend])
# prior_inde_err = np.mean(prior_inde_rmse[iobsbeg - 1: iobsend])
analy_err = np.mean(analy_rmse[iobsbeg - 1: iobsend])
# analy_inde_err = np.mean(analy_inde_rmse[iobsbeg - 1: iobsend])
analy_indeself_err = np.mean(analy_indeself_rmse[iobsbeg - 1: iobsend])
prior_spread = np.mean(prior_spread_rmse[iobsbeg - 1: iobsend])
prior_spread_inde = np.mean(prior_spread_inde_rmse[iobsbeg - 1: iobsend])
analy_spread = np.mean(analy_spread_rmse[iobsbeg - 1: iobsend])
analy_spread_inde = np.mean(analy_spread_inde_rmse[iobsbeg - 1: iobsend])

# print('prior error = {0:.6f}'.format(prior_err))
print('analysis error = {0:.6f}'.format(analy_err))
# print('prior_inde error = {0:.6f}'.format(prior_inde_err))
print('analy_inde_self error = {0:.6f}'.format(analy_indeself_err))
# print('analy_inde error = {0:.6f}'.format(analy_inde_err))
print('prior_spread = {0:.6f}'.format(prior_spread))
print('prior_spread_inde = {0:.6f}'.format(prior_spread_inde))
print('analy_spread = {0:.6f}'.format(analy_spread))
print('analy_spread_inde = {0:.6f}'.format(analy_spread_inde))
```
