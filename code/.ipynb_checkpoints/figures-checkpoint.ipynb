```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
# import sys 
# sys.path.append('/home/lllei/AI_localization/L05/git_repo/general')
# sys.path.append('/home/lllei/AI_localization/NAFNet/')
# from construct_GC_2d import construct_GC_2d
# from construct_func_2d import construct_func_2d
# import torch
# from basicsr.models import create_model
# from basicsr.utils.options import parse
```

```python
model_size = 960
obs_density = 4
model_grids = np.arange(1, model_size + 1)
obs_grids = model_grids[model_grids % obs_density == 0]

# draw snapshots of gain matrices
def draw_s(colorlim,fname=''):
    kg_f = np.load('{0:}_40.npy'.format(fname))
    kg_t = np.load('{0:}_2000.npy'.format(fname))
    kg_gc = np.load('{0:}_GC.npy'.format(fname))
    kg_lrlf = np.load('{0:}_LRLF.npy'.format(fname))
    kg_clf = np.load('{0:}_CLF.npy'.format(fname))
    kg_elf = np.load('{0:}_ELF.npy'.format(fname))
    kg_celf = np.load('{0:}_CELF.npy'.format(fname))

    images = []
    cmap_op = plt.cm.seismic
    nax=7
    fig, axs = plt.subplots(1,nax)
    axs[0].set_title('40',fontsize=10)
    images.append(axs[0].imshow(kg_f, cmap=cmap_op))

    axs[1].set_title('2000',fontsize=10)
    images.append(axs[1].imshow(kg_t, cmap=cmap_op))

    axs[2].set_title('40+GC',fontsize=10)
    images.append(axs[2].imshow(kg_gc, cmap=cmap_op))

    axs[3].set_title('40+LRLF',fontsize=10)
    images.append(axs[3].imshow(kg_lrlf, cmap=cmap_op))

    axs[4].set_title('40+CLF',fontsize=10)
    images.append(axs[4].imshow(kg_clf, cmap=cmap_op))

    axs[5].set_title('40+ELF',fontsize=10)
    images.append(axs[5].imshow(kg_elf, cmap=cmap_op))

    axs[6].set_title('40+CELF',fontsize=10)
    images.append(axs[6].imshow(kg_celf, cmap=cmap_op))

    for i in range(1,nax):
        axs[i].yaxis.set_ticklabels([])

    vmin = min(image.get_array().min() for image in images)
    vmax = max(image.get_array().max() for image in images)
    norm = colors.Normalize(vmin=-colorlim, vmax=colorlim)
    for im in images:
        im.set_norm(norm)

    cbar = fig.colorbar(images[-1], ax=axs[0:], orientation='vertical', fraction=.023)

    fig.subplots_adjust(bottom=0.1, top=0.9, left=0.1, right=0.86,
                        wspace=0.2, hspace=0.05)
    plt.show()
    
    
# draw time-average of gain matrices
def draw_s_tmean(colorlim,fname=''):
    kg_f = np.load('{0:}_40.npy'.format(fname))
    kg_t = np.load('{0:}_2000.npy'.format(fname))
    kg_gc = np.load('{0:}_GC.npy'.format(fname))
    kg_lrlf = np.load('{0:}_LRLF.npy'.format(fname))
    kg_clf = np.load('{0:}_CLF.npy'.format(fname))
    kg_elf = np.load('{0:}_ELF.npy'.format(fname))
    kg_celf = np.load('{0:}_CELF.npy'.format(fname))

    images = []
    cmap_op = plt.cm.seismic
    nax=7
    fig, axs = plt.subplots(1,nax)
    axs[0].set_title('40',fontsize=10)
    images.append(axs[0].imshow(kg_f, cmap=cmap_op))

    axs[1].set_title('2000',fontsize=10)
    images.append(axs[1].imshow(kg_t, cmap=cmap_op))

    axs[2].set_title('40+GC',fontsize=10)
    images.append(axs[2].imshow(kg_gc, cmap=cmap_op))

    axs[3].set_title('40+LRLF',fontsize=10)
    images.append(axs[3].imshow(kg_lrlf, cmap=cmap_op))

    axs[4].set_title('40+CLF',fontsize=10)
    images.append(axs[4].imshow(kg_clf, cmap=cmap_op))

    axs[5].set_title('40+ELF',fontsize=10)
    images.append(axs[5].imshow(kg_elf, cmap=cmap_op))

    axs[6].set_title('40+CELF',fontsize=10)
    images.append(axs[6].imshow(kg_celf, cmap=cmap_op))
    
    for i in range(1,nax):
        axs[i].yaxis.set_ticklabels([])

    vmin = min(image.get_array().min() for image in images)
    vmax = max(image.get_array().max() for image in images)
    norm = colors.Normalize(vmin=-colorlim, vmax=colorlim)
    for im in images:
        im.set_norm(norm)

    cbar = fig.colorbar(images[-1], ax=axs[0:], orientation='vertical', fraction=.023)

    fig.subplots_adjust(bottom=0.1, top=0.9, left=0.1, right=0.86,
                        wspace=0.2, hspace=0.05)
    
    plt.show()

    
# draw time evolutions of gain matrices
def draw_t(fname=''):
    kg_f = np.load('{0:}_40.npy'.format(fname))
    kg_t = np.load('{0:}_2000.npy'.format(fname))
    kg_gc = np.load('{0:}_GC.npy'.format(fname))
    kg_lrlf = np.load('{0:}_LRLF.npy'.format(fname))
    kg_clf = np.load('{0:}_CLF.npy'.format(fname))
    kg_elf = np.load('{0:}_ELF.npy'.format(fname))
    kg_celf = np.load('{0:}_CELF.npy'.format(fname))

    yloc = 120
    nrow = 7
    ncol = 6
    
    plt.rcParams.update({'font.size': 8})

    fig, axes = plt.subplots(nrow, ncol,sharex=True,sharey=True)
    images = []
    xtick = range(-480,480)
    
    for n in range(ncol):
        axes[0,n].plot(xtick,kg_f[n,:],'b',linewidth=1)
        axes[1,n].plot(xtick,kg_t[n,:],'b',linewidth=1)
        axes[2,n].plot(xtick,kg_gc[n,:],'b',linewidth=1)
        axes[3,n].plot(xtick,kg_lrlf[n,:],'b',linewidth=1)
        axes[4,n].plot(xtick,kg_clf[n,:],'b',linewidth=1)
        axes[5,n].plot(xtick,kg_elf[n,:],'b',linewidth=1)
        axes[6,n].plot(xtick,kg_celf[n,:],'b',linewidth=1)

        for i in range(nrow):
            axes[i,n].set_xlim(-480, 480)
            axes[i,n].set_xticklabels(np.arange(-480, 480, step=240), fontsize=8)

    for n in range(ncol):
        axes[0,n].set_title('{:d}d'.format(n))

    axs = axes[0,-1].twinx()
    axs.set_ylabel('40',rotation='horizontal')
    axs.yaxis.set_label_coords(1.3, 0.5)
    axs.set_yticks([])

    axs = axes[1,-1].twinx()
    axs.set_ylabel('2000',rotation='horizontal')
    axs.yaxis.set_label_coords(1.35, 0.5)
    axs.set_yticks([])

    axs = axes[2,-1].twinx()
    axs.set_ylabel('40+GC',rotation='horizontal')
    axs.yaxis.set_label_coords(1.4, 0.5)
    axs.set_yticks([])

    axs = axes[3,-1].twinx()
    axs.set_ylabel('40+LRLF',rotation='horizontal')
    axs.yaxis.set_label_coords(1.45, 0.5)
    axs.set_yticks([])

    axs = axes[4,-1].twinx()
    axs.set_ylabel('40+CLF',rotation='horizontal')
    axs.yaxis.set_label_coords(1.4, 0.5)
    axs.set_yticks([])

    axs = axes[5,-1].twinx()
    axs.set_ylabel('40+ELF',rotation='horizontal')
    axs.yaxis.set_label_coords(1.4, 0.5)
    axs.set_yticks([])

    axs = axes[6,-1].twinx()
    axs.set_ylabel('40+CELF',rotation='horizontal')
    axs.yaxis.set_label_coords(1.45, 0.5)
    axs.set_yticks([])

    fig.subplots_adjust(bottom=0.14, top=0.9, left=0.14, right=0.9,
                        wspace=0.06, hspace=0.06)
    fig.text(0.5, 0.04, 'displacement', ha='center')
    fig.text(0.04, 0.5, 'gain', va='center', rotation='vertical')
    plt.show()
```
